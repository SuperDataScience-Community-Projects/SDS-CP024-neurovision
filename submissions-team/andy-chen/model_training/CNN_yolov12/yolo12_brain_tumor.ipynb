```python
import cv2
import torch
from ultralytics import YOLO

# Load the trained model
best_yolov12_model = YOLO("yolov12/runs/detect/train3/weights/best.pt")

# Path to your test image
image_path = "datasets/brain-tumor/valid/images/val_1 (1).jpg"

# Run prediction
results = best_yolov12_model(image_path)

# Read the image
image = cv2.imread(image_path)

# Process each detection
for result in results:
    # Get the bounding boxes, confidence scores, and class labels
    boxes = result.boxes.xyxy.cpu().numpy()  # get boxes on cpu in xy format
    confidences = result.boxes.conf.cpu().numpy()
    class_ids = result.boxes.cls.cpu().numpy()

    # Draw bounding boxes and add labels
    for box, conf, cls in zip(boxes, confidences, class_ids):
        # Convert coordinates to integers
        x1, y1, x2, y2 = map(int, box)
        
        # Get class name
        class_name = best_yolov12_model.names[int(cls)]
        
        # Draw bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        
        # Create label with class and confidence
        label = f"{class_name}: {conf:.2f}"
        
        # Put text on the image
        cv2.putText(image, label, (x1, y1 - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Display the image with predictions
cv2.imshow("Brain Tumor Detection", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Optionally, save the annotated image
cv2.imwrite("prediction_result.jpg", image)

# Print out detailed results
print("Prediction Results:")
for result in results:
    print(f"Boxes: {result.boxes}")
    print(f"Confidence Scores: {result.boxes.conf}")
    print(f"Class IDs: {result.boxes.cls}")
```

```python
import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the trained model
best_yolov12_model = YOLO("yolov12/runs/detect/train3/weights/best.pt")

# Path to your test image
image_path = "datasets/brain-tumor/valid/images/val_1 (1).jpg"

# Run prediction
results = best_yolov12_model(image_path)

# Read the image using OpenCV
image = cv2.imread(image_path)

# Check if image is loaded
if image is None:
    print(f"Error: Could not read the image at {image_path}")
    exit()

# Ensure the image is not empty
print(f"Image shape: {image.shape}")

# Process each detection
for result in results:
    # Get the bounding boxes, confidence scores, and class labels
    boxes = result.boxes.xyxy.cpu().numpy()  # get boxes on cpu in xy format
    confidences = result.boxes.conf.cpu().numpy()
    class_ids = result.boxes.cls.cpu().numpy()

    print(f"Number of detections: {len(boxes)}")

    # Draw bounding boxes and add labels
    for box, conf, cls in zip(boxes, confidences, class_ids):
        # Convert coordinates to integers
        x1, y1, x2, y2 = map(int, box)
        
        # Get class name
        class_name = best_yolov12_model.names[int(cls)]
        
        # Draw bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        
        # Create label with class and confidence
        label = f"{class_name}: {conf:.2f}"
        
        # Put text on the image
        cv2.putText(image, label, (x1, y1 - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Use matplotlib for display (alternative to cv2.imshow)
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('on')
plt.title('Brain Tumor Detection Results')
plt.tight_layout()
plt.show()

# Optionally, save the annotated image
cv2.imwrite("prediction_result.jpg", image)

# Print out detailed results
print("Prediction Results:")
for result in results:
    print(f"Boxes: {result.boxes}")
    print(f"Confidence Scores: {result.boxes.conf}")
    print(f"Class IDs: {result.boxes.cls}")
```
