```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
config_file = '/content/drive/MyDrive/brain_tumor_dataset/config.yaml'
```

```python
from ultralytics import YOLO
model_name = 'yolov8s'
model = YOLO(f"{model_name}.pt")

results = model.train(
    data = config_file,
    project="/content/drive/MyDrive/Brain_Tumor_Training",
    name = model_name,
    epochs = 25,
    imgsz = 640,
    batch = 16,
    lr0=0.01,
    dropout= 0.15
)
```

```python
print(f"Precision: {results.box.p}")
print(f"Recall: {results.box.r}")
print(f"F1-score: {results.box.f1}")
```
